```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel('Global Superstore Orders 2016.xlsx')
df.rename(columns={'Order Date': 'Order_Date', 'Ship Date': 'Ship_Date'}, inplace=True)
df.head()
```

```python
df['Order_Date'] = df['Order_Date'] + pd.offsets.DateOffset(years=5)
df['Ship_Date'] = df['Ship_Date'] + pd.offsets.DateOffset(years=5)
df
```

```python
pin_date = max(df.Order_Date) + dt.timedelta(1)
df['TotalSum'] = df.Quantity * df.Sales
df.head()
```

```python
rfm = df.groupby('Customer ID').agg({
    'Order_Date' : lambda x: (pin_date - x.max()).days,
    'Order ID' : 'count', 
    'TotalSum' : 'sum'})
rfm.rename(columns = {'Order_Date' : 'Recency',
                      'Order ID' : 'Frequency', 
                      'TotalSum' : 'Monetary'}, inplace = True)
rfm.head()
```

```python
rfm['Frequency'] = rfm['Frequency'].astype(float)

r_labels = range(4, 0, -1)
r_groups = pd.qcut(rfm.Recency, q = 4, labels = r_labels)

m_labels = range(1, 5)
m_groups = pd.qcut(rfm.Monetary, q = 4, labels = m_labels)
```

```python
def pct_rank_qcut(series, n):
    edges = pd.Series([float(i) / n for i in range(n + 1)])
    f = lambda x: (edges >= x).argmax()
    return series.rank(pct=1).apply(f)

rfm['F'] = pct_rank_qcut(rfm.Frequency, 4)
rfm['R'] = r_groups.values
rfm['M'] = m_groups.values# sum up the three columns
rfm['RFM_Segment'] = rfm.apply(lambda x: str(x['R']) + str(x['F']) + str(x['M']), axis = 1)
rfm['RFM_Score'] = rfm[['R', 'F', 'M']].sum(axis = 1)
rfm.head()
```

```python
score_labels = ['Green', 'Bronze', 'Silver', 'Gold']
score_groups = pd.qcut(rfm.RFM_Score, q = 4, labels = score_labels)
rfm['RFM_Level'] = score_groups.values
rfm.head()
```

```python
rfm.to_excel('RFM_Segment.xlsx')
```

```python
total_avg = rfm.iloc[:, 0:3].mean()
cluster_avg = rfm.groupby('RFM_Level').mean().iloc[:, 0:3]
prop_rfm = cluster_avg/total_avg - 1
sns.heatmap(prop_rfm, cmap= 'Oranges', fmt= '.2f', annot = True)
plt.title('Heatmap')
plt.plot()
```

```python
rfm_g = rfm.groupby('RFM_Level').count()
color_list = ['green', 'peru','gainsboro','gold']

plt.bar(rfm_g.index, rfm_g.F,color = color_list)
plt.xticks(ticks=rfm_g.index, labels=rfm_g.index)
plt.xlabel('Class')
plt.ylabel('Customer Count')
plt.title('RFM Customer Count')
plt.show()
```
